# 단순 선형 회귀 - 당뇨병 사례

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### -당뇨병데이터

```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df['target'] = diabetes.target
df.head()
```

```python
df.describe()
```

### - 당뇨병(target) vs. BMI

```python
X = df.bmi.values.reshape(-1,1)
y = diabetes.target
X.shape, y.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X,y)
```

```python
weight, bias = lr.coef_, lr.intercept_
rss = lr._residues
weight[0], bias, rss
```

```python
lr.score(X,y)
#R square
```

### -train/test data분리하고
### -test size= 0.1

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, random_state=2021
)
#분류값이 아니니까 stratify는 적용no
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
pred = lr.predict(X_test)
pred[:4]
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, pred)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test,pred)
rmse =np.sqrt(mse)
mse, rmse
```

### - 그래프로 확인

```python
plt.scatter(X, y)
plt.grid()
plt.show
```

```python
w , b = lr.coef_, lr.intercept_
plt.scatter(X_train, y_train)
plt.plot([-0.1,0.18],[-0.1*w+b, 0.18*w+b],'r',lw=3)
plt.grid()
plt.show
```

```python
plt.scatter(X_test, y_test)
plt.plot([-0.1,0.18],[-0.1*w+b, 0.18*w+b],'r',lw=3)
plt.grid()
plt.show
```

### - 회귀 표본 만들어 연습하기

```python
from sklearn.datasets import make_regression
X, y = make_regression(n_samples=100, n_features=1,n_targets=1,random_state=2021)
```

```python
plt.scatter(X, y)
plt.grid()
plt.show
```

```python
X, y = make_regression(n_samples=100, n_features=1,n_targets=1,noise = 100, random_state=2021)
plt.scatter(X, y)
plt.grid()
plt.show
# noise = 100
```

```python
X, y = make_regression(n_samples=100, n_features=1,n_targets=1,noise = 100,bias=10, random_state=2021)
plt.scatter(X, y)
plt.grid()
plt.show
```

```python
lr = LinearRegression()
lr.fit(X,y)
w, b = lr.coef_[0], lr.intercept_
```

```python
plt.scatter(X, y)
plt.plot([-3,4],[-3*w+b, 4*w+b],'r',lw=3)
plt.grid()
plt.show
```

```python
r2 = lr.score(X,y)
w, b, r2
```
